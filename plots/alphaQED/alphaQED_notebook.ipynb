```python
# import pandas
import pandas as pd
# import matplotlib
import matplotlib.pyplot as plt
```

```python
alphaQED_data = pd.read_csv("alphaQED.txt", sep = '\t')
Q2 = alphaQED_data["Q2"]
Inv_AlphaQED = alphaQED_data["InvalphaQED"]
```

```python
plt.plot(Q2, Inv_AlphaQED)
plt.xlabel(r'$Q^2 (GeV^2)$')
plt.xscale("log")
plt.ylabel(r'$\alpha^{-1}$')
plt.ylim(122, 138)
plt.savefig("alphaQED_vs_Q2.pdf")
plt.show()
```
