```python
# import numpy, pyplot and pandas
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
hard_pomeron_data = pd.read_csv("SigmaGammaProton_HardPomeron_pred.txt", sep = '\t')
soft_pomeron_data = pd.read_csv("SigmaGammaProton_SoftPomeron_pred.txt", sep = '\t')
exp_data = pd.read_csv("SigmaGammaProton.txt", sep = '\t')
```

```python
# We can plot the data to check quickly if we have done it correctly
plt.errorbar(exp_data["W(GeV)"], exp_data["sigma(mub)"], xerr = [exp_data["WErrorm"],exp_data["WErrorp"]], 
             yerr = [exp_data["Errorm"],exp_data["Errorp"]], fmt = "o", label = "exp. data")
plt.plot(hard_pomeron_data["W"], hard_pomeron_data["Pred"], label = "Hard Pomeron")
plt.plot(soft_pomeron_data["W"], soft_pomeron_data["Pred"], label = "Soft Pomeron")
plt.ylabel(r'$\sigma(\mu b)$')
plt.xlabel(r'$\sqrt{s}(GeV)$')
plt.xscale("log")
plt.title(r'$\sigma$'+" vs " + r'$\sqrt{s}$')
plt.legend()
plt.show()
```
