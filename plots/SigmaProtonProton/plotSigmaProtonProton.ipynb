```python
# import numpy, pandas and matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
SigmaPP_pred = pd.read_csv("SigmaProtonProton_HardPomeron_pred.txt",sep = '\t')
exp_data = pd.read_csv("SigmaProtonProton_data.txt", sep = '\t')
```

```python
# We can plot the data to check quickly if we have done it correctly
plt.errorbar(exp_data["W(GeV)"], exp_data["sigma(mb)"], xerr = [exp_data["WErrorm"],exp_data["WErrorp"]], 
             yerr = [exp_data["Errorm"],exp_data["Errorp"]], fmt = "o", label = "exp. data")
plt.plot(SigmaPP_pred["W"], SigmaPP_pred["Pred"], label = "Hard Pomeron")
plt.ylabel(r'$\sigma(mb)$')
plt.xlabel(r'$\sqrt{s}(GeV)$')
plt.xscale("log")
plt.title(r'$\sigma(p p \rightarrow X)$'+" vs " + r'$\sqrt{s}$')
plt.legend()
plt.savefig("SigmaProtonProton_HardPomeron.pdf")
plt.show()
```
