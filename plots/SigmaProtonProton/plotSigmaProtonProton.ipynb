```python
# import numpy, pandas and matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
SigmaPP_pred_HardPomeron_W_lt_386 = pd.read_csv("SigmaProtonProton_pred_HardPomeron_W_lt_386.txt",sep = '\t')
SigmaPP_pred_HardPomeron_W_lt_10000 = pd.read_csv("SigmaProtonProton_pred_HardPomeron_W_lt_10000.txt",sep = '\t')
SigmaPP_pred_HardPomeron = pd.read_csv("SigmaProtonProton_pred_HardPomeron.txt", sep = '\t')
exp_data_W_lt_386 = pd.read_csv("SigmaProtonProton_data_W_lt_386.txt", sep = '\t')
exp_data_W_lt_10000 = pd.read_csv("SigmaProtonProton_data_W_lt_10000.txt", sep = '\t')
exp_data = pd.read_csv("SigmaProtonProton_data.txt", sep = '\t')
```

```python
def plotExpDataVsPred(exp_data, pred_data, label, sigma_min, sigma_max, output_file):
    """
        Makes a plot where the experimental points can be compared
        with the predictions of the best fit to those points.
        exp_data : Pandas data frame with the experimental points
        pred_data : Pandas data frame with the best fit predicted points
        label : information about the Pomeron kernel used in the fit
        output_file : name of the file of the saved plot
    """
    plt.errorbar(exp_data["W(GeV)"], exp_data["sigma(mb)"], 
                 xerr = [exp_data["WErrorm"], exp_data["WErrorp"]], 
                 yerr = [exp_data["Errorm"], exp_data["Errorp"]], fmt = "o", label = "exp. data")
    plt.plot(pred_data["W"], pred_data["Pred"], label = label)
    plt.ylabel(r'$\sigma(mb)$')
    plt.xlabel(r'$\sqrt{s}(GeV)$')
    plt.xscale("log")
    plt.title(r'$\sigma$'+" vs " + r'$\sqrt{s}$')
    plt.legend()
    plt.ylim(sigma_min, sigma_max)
    plt.savefig(output_file)
    plt.show()
```

```python
# Plot results of the fit sqrt(s) < 386
plotExpDataVsPred(exp_data_W_lt_386, SigmaPP_pred_HardPomeron_W_lt_386, "Hard Pomeron", 20, 100, "SigmaProtonProton_HardPomeron_W_lt_386.pdf")
```

```python
# Plot results of the fit sqrt(s) < 10000
plotExpDataVsPred(exp_data_W_lt_10000, SigmaPP_pred_HardPomeron_W_lt_10000, "Hard Pomeron",30, 150, "SigmaProtonProton_HardPomeron_W_lt_10000.pdf")
```

```python
# Plot results of the fit sqrt(s) > 3 GeV
plotExpDataVsPred(exp_data, SigmaPP_pred_HardPomeron, "Hard Pomeron",30, 200, "SigmaProtonProton_HardPomeron.pdf")
```
